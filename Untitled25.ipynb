```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
dept_df = pd.DataFrame({
    'FISCAL YEAR': [2022, 2022, 2023, 2023, 2024, 2024],
    'ALL FUNDS': [100000, 150000, 120000, 160000, 130000, 170000],
    'CITY FUND': [80000, 130000, 100000, 140000, 110000, 150000]
})

# Calculate the variance between 'ALL FUNDS' and 'CITY FUND'
dept_df['Variance'] = dept_df['ALL FUNDS'] - dept_df['CITY FUND']

# Get top 10 variances
top_10_variances = dept_df.nlargest(10, 'Variance')

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(top_10_variances.index, top_10_variances['Variance'], color='skyblue')

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Variance')
plt.title('Top 10 Variances between ALL FUNDS and CITY FUND')
plt.xticks(top_10_variances.index, [f"{row['FISCAL YEAR']} - {row['ALL FUNDS']} vs {row['CITY FUND']}" for _, row in top_10_variances.iterrows()], rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
dept_df = pd.DataFrame({
    'FISCAL YEAR': [2022, 2022, 2023, 2023, 2024, 2024],
    'ALL FUNDS': [100000, 150000, 120000, 160000, 130000, 170000],
    'CITY FUND': [80000, 130000, 100000, 140000, 110000, 150000]
})

# Calculate the variance between 'ALL FUNDS' and 'CITY FUND'
dept_df['Variance'] = dept_df['ALL FUNDS'] - dept_df['CITY FUND']

# Sort by 'FISCAL YEAR' for a line plot
sorted_df = dept_df.sort_values(by='FISCAL YEAR')

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(sorted_df['FISCAL YEAR'], sorted_df['Variance'], marker='o', linestyle='-', color='b')

# Adding labels and title
plt.xlabel('Fiscal Year')
plt.ylabel('Variance')
plt.title('Variance Trends Over Fiscal Years')
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
dept_df = pd.DataFrame({
    'FISCAL YEAR': [2022, 2022, 2023, 2023, 2024, 2024],
    'ALL FUNDS': [100000, 150000, 120000, 160000, 130000, 170000],
    'CITY FUND': [80000, 130000, 100000, 140000, 110000, 150000]
})

# Calculate the variance between 'ALL FUNDS' and 'CITY FUND'
dept_df['Variance'] = dept_df['ALL FUNDS'] - dept_df['CITY FUND']

# Plotting
plt.figure(figsize=(12, 8))
plt.boxplot(dept_df['Variance'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Adding labels and title
plt.xlabel('Variance')
plt.title('Box Plot of Variance Values')

# Display the plot
plt.tight_layout()
plt.show()
```
